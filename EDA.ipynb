```python
def get_highly_correlated_cols(df):
    
    correlation_df = df[df.columns].corr()
    #row,col,tuple_arr,correlations = 1,0,[],[]
    tuple_arr,correlations = [],[]
    
    for row in range(len(correlation_df)):
        for col in range(row):
            if(correlation_df.iloc[row,col] > 0.6):
                correlations.append(correlation_df.iloc[row,col])
                tuple_arr.append((row,col))
                
    return correlations , tuple_arr , correlation_df
```

```python
def print_correlated(df,nameOfNeighborhood):
    for i in range(len(correlations)):        
        print("corr of %s('%s', '%s')=%f" % (nameOfNeighborhood,df.columns[tuple_arr[i][1]+1], df.columns[tuple_arr[i][0]+1], correlations[i]))
```

```python
correlations, tuple_arr, correlation_RovaBet = get_highly_correlated_cols(RovaBet)
print_correlated(RovaBet,'Rova Bet')
correlation_RovaBet
```

```python
correlations, tuple_arr, correlation_RovaDaled = get_highly_correlated_cols(RovaDaled)
print_correlated(RovaDaled,'Rova Daled')
correlation_RovaDaled
```

```python
correlations, tuple_arr, correlation_RovaYudAlef = get_highly_correlated_cols(RovaYudAlef)
print_correlated(RovaYudAlef,'Rova Yud Alef')
correlation_RovaYudAlef
```

```python
correlations, tuple_arr, correlation_RovaVav = get_highly_correlated_cols(RovaVav)
print_correlated(RovaVav,'Rova Vav')
correlation_RovaVav
```
